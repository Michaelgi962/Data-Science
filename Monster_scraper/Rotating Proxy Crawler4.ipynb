```python
Https_txt_file = 'proxyLists/https_proxys_part_3.txt'
#Download New HTTPS IPs .txt file from https://www.proxy-list.download/HTTPS
https_proxys = pd.read_csv(Https_txt_file, sep=" ", header=None)
#Convert to array
https_proxys = https_proxys[0].values
#Convert to list
https_proxys = list(https_proxys)

Http_txt_file = 'proxyLists/http_proxys_part_3.txt'
#Download New HTTPS IPs .txt file from https://www.proxy-list.download/HTTP
http_proxys = pd.read_csv(Http_txt_file, sep=" ", header=None)
#Convert to array
http_proxys = http_proxys[0].values
#Convert to list
http_proxys = list(http_proxys)
```

```python
header = {}
def randHeader():
    ua = UserAgent()
    header['User-Agent'] = ua.random
```

```python
# Retrieve a random index for the https proxy
def randomIndexHttps():
    num = random.randint(0, len(https_proxys) - 1)
    return num

# Retrieve a random index for the http proxy
def randomIndexHttp():
    num = random.randint(0, len(http_proxys) - 1)
    return num
```

```python
proxyDict = {}
def randProx():
    # Choose a random proxy ip
    proxy_index_http = randomIndexHttp()
    proxy_index_https = randomIndexHttps()
   
    http_proxy = http_proxys[proxy_index_http]
    https_proxy = https_proxys[proxy_index_https]
    
    http_proxy_final  = 'http://' + http_proxy
    https_proxy_final = 'https://' + https_proxy

    proxyDict['http'] = http_proxy_final
    proxyDict['https'] = https_proxy_final
    
    return proxy_index_https,proxy_index_http
```

```python
url = 'http://icanhazip.com'
if url[0:5] == 'http:':
    for n in range(len(http_proxys)-1):
        # Every 10 requests, generate a new proxy
        t1 = time.time()
        # Choose a random proxy ip
        proxy_index_https,proxy_index_http = randProx()
        print(len(https_proxys),len(http_proxys))
        # Make the call
        try:
            my_ip = requests.get(url,headers=header,proxies=proxyDict,timeout=1).text
            t2 = time.time()
            t = t2-t1
            print('time:'+str(t)+', #' + str(n) + ': ' + str(proxyDict['http']))

        except: # If error, delete this proxy and find another one    
            t2 = time.time()
            del http_proxys[proxy_index_http]
            print('time:'+str(t2-t1)+str(proxyDict['http'])+ ' deleted.')

elif url[0:5] == 'https':
    for n in range(len(https_proxys)-1):
        # Every 10 requests, generate a new proxy
        t1 = time.time()
        # Choose a random proxy ip
        proxy_index_https,proxy_index_http = randProx()
        print(len(https_proxys),len(http_proxys))
        # Make the call
        try:
            my_ip = requests.get(url,headers=header,proxies=proxyDict,timeout=2).text
            t2 = time.time()
            t = t2-t1
            print('time:'+str(t)+', #' + str(n) + ': ' + str(proxyDict['https']))

        except: # If error, delete this proxy and find another one                 
            t2 = time.time()
            del https_proxys[proxy_index_https]
            print('time:'+str(t2-t1)+str(proxyDict['https'])+ ' deleted.')
```
